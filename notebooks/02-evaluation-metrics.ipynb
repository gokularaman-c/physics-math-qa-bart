```python
import torch
from transformers import BartForConditionalGeneration, BartTokenizer

model_path = "/kaggle/input/phymath-qa/phymath_qa/bart_qa_finetuned"  
tokenizer = BartTokenizer.from_pretrained(model_path)
model = BartForConditionalGeneration.from_pretrained(model_path)
model.eval()

device = "cuda" if torch.cuda.is_available() else "cpu"
model.to(device)

def answer_question(question, context):#bart 
    input_text = f"{question} {context}"
    inputs = tokenizer(input_text, return_tensors="pt", truncation=True, padding=True, max_length=1024)
    inputs = {k: v.to(device) for k, v in inputs.items()}
    
    outputs = model.generate(
        **inputs,
        max_length=250,
        num_beams=5,
        forced_bos_token_id=tokenizer.bos_token_id
    )
    predicted_answer = tokenizer.decode(outputs[0], skip_special_tokens=True)
    
    return predicted_answer
```

```python
from evaluate import load

rouge = load("rouge")
bleu = load("bleu")
squad_metric = load("squad")

predictions = data["preds"].tolist()
references = data["answers"].tolist()
```

```python
rouge_score = rouge.compute(predictions=predictions, references=references)
bleu_score = bleu.compute(predictions=predictions, references=[[ref] for ref in references])

squad_predictions = [{"id": str(i), "prediction_text": pred} for i, pred in enumerate(predictions)]
squad_references = [{"id": str(i), "answers": {"text": [ref], "answer_start": [0]}} for i, ref in enumerate(references)]
em_score = squad_metric.compute(predictions=squad_predictions, references=squad_references)
```
